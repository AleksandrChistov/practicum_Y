- ponds_nearest — расстояние до ближайшего водоёма (м)

- studio — квартира-студия (булев тип)

```python
#Поменяем последовательность столбцов на более последовательную по важности информации
df = df[[
 'locality_name', 
 'living_area',
 'kitchen_area',
 'total_area',
 'rooms',
 'ceiling_height',
 'last_price',
 'balcony',
 'floor',
 'floors_total',
 'is_apartment',
 'studio',
 'open_plan',
 'first_day_exposition',
 'days_exposition',  
 'cityCenters_nearest',
 'airports_nearest',
 'parks_around3000',
 'parks_nearest', 
 'ponds_around3000',
 'ponds_nearest',
 'total_images',
]]
```

```python
 #Посмотрим, как получилось
df.head()
```

Сейчас разберемся с пропусками и типом данных

```python
#заменим пропуски в balcony на 0, согласно вводным условиям и изменим тип данных на int
df['balcony'] = df['balcony'].fillna(0).astype('Int64')
```

```python
#Дату публикации приведем в формат год-месяц-день, убрав лишнюю информацию
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format="%Y-%m-%d")
```

```python
df.head()
```

```python
#столбец апартаменты видимо появился не вместе с началом ведения записей, все пропущенные значения заменим на False, 
#апартаменты появились относительно недавно
df['is_apartment'] = df['is_apartment'].fillna(False)
```

Изменим тип данных в столбцах, где не нужна точность после запятой

```python
astype_cols = [col for col in df.columns if col in ['floors_total', 'last_price', 'parks_around3000', 'ponds_around3000']]

for col in astype_cols:
    df[col] = df[col].astype('Int64')
```

```python
df.info()
```

```python
#Сколько пропусков 'kitchen_area'
df['kitchen_area'].isnull().sum()
```

```python
#Зная 'total_area' и 'living_area', узнаем их разность и заполним 'kitchen_area'
df.loc[(df['living_area'] > 0 & (df['kitchen_area'].isnull() == True)), 'kitchen_area'] = df['total_area'] - df['living_area']
```

```python
#Посмотрим колько пропусков 'kitchen_area' нам удалось заполнить
df['kitchen_area'].isnull().sum()
```

Мы заполнили 30% пропущенных значений. Аналогично поступим с заполнением 'living_area'

```python
#Сколько пропусков 'living_area'
df['living_area'].isnull().sum()
```

```python
#Зная 'total_area' и 'living_area', узнаем их разность и заполним 'kitchen_area'
df.loc[(df['kitchen_area'] > 0 & (df['living_area'].isnull() == True)), 'living_area'] = df['total_area'] - df['kitchen_area']
```

```python
#Посмотрим сколько пропусков 'kitchen_area' осталось после замены
df['living_area'].isnull().sum()
```

Высота потолков. Данная графа не заполнена у 39% объектов недвижимости, при этом у заполненных цифр есть довольно нереальные значения. Посмотрим 

```python
df['ceiling_height'].sort_values().unique()
```

Высота потолков до 2м кажется слишком маленькой, чтобы быть правдой, но возможно это какие-то старые постройки.
Высота потолков больше 4-5м кажется уже слишком большой и больше подходит для коммерческих объектов.  
Высота потолков 20+ м больше похожа на человеческую ошибку при заполнении, здесь можно скорректировать данные.
Значения высоты 1, 1.2 и 100 метров являются аномалией.

```python
df[df['ceiling_height'] <= 2]
```

```python
df[df['ceiling_height'] >= 5]
```

Очень похоже, что при заполнении ошиблить с разрядом числа и не поставили разделительную точку. исправим

```python
df.loc[(df["ceiling_height"] >= 20), "ceiling_height"] = df["ceiling_height"]/10
```

```python
#Проверим, как измеились данные
df[df['ceiling_height'] >= 5]
```

Теперь заполним пропущенные значения высоты потолков.  
Если бы мы знали типовой проект дома, то можно было бы посмотреть высоту в аналогичных квартирах в данных проектах.  
Нам же придется заполнять медианными значениями.

```python
ceiling_medians = df.pivot_table(index='floors_total', values='ceiling_height', aggfunc=['median', 'count'])
ceiling_medians.columns = ['median', 'count']
ceiling_medians.head()
```

Медианные значения вставляем на места пропущенных значений высоты потолков домов соответствующей этажности


```python
#Пишем функцию замены пропущенных значений на медианы
for floor in ceiling_medians.index:
    df.loc[(df['ceiling_height'].isna()) & 
             (df['floors_total'] == floor), 
             'ceiling_height'] = ceiling_medians.loc[floor, 'median']
```

```python
#В домах где этажность неизвестна, заменяем значения высоты потолков на общую медиану, 
for floor in ceiling_medians.index:
    df.loc[(df['floors_total'].isna()) | 
             ((df['floors_total'] == floor) & (ceiling_medians.loc[floor, 'count'] == 0)), 
             'ceiling_height'] = df['ceiling_height']
```

```python
#Проверяем, как прошла замена
df['ceiling_height'].isnull().sum()
```

Ещё остались пропуски, посмотрим внимательнее на остальные столбцы в этих строках

```python
df.query('ceiling_height .isnull() == True')
```

В этих строках также отсутствуют данные о количестве этажей. Удалим их, чтобы результаты анализа были максимально корректными.

```python
df = df.loc[~((df['ceiling_height'].isnull() == True)),:]
```

```python
#Проверяем, как удалились
df['ceiling_height'].isnull().sum()
```

Посмотрим количество строк без указания населенного пункта. 

```python
df['locality_name'].isnull().sum()
```

```python
#Удалим их, так они могут исказить наше иследование
df.dropna(subset=['locality_name'], inplace=True)
```

## Расчёты и добавление результатов в таблицу

Рассчитаем цену квадратного метра и для удобства округлим

```python
df['price_per_meter'] = round(df['last_price'] / df['total_area']).astype('int64')
```

Добавим столбцы дня недели, месяца и года публикации. Посмотрим результат

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
df.head(10)
```

```python
#Напишем функцию категоризации по этажам, используя метод apply и применим её
def floor_definition(df):
    if df['floor'] == 1:
        return 'первый'
    elif (df['floors_total'] == df['floor']) & (df['floor'] != 1):
        return 'последний'
    else:
        return 'другой'
df['floor_cat'] = df.apply(floor_definition, axis = 1)
```

Добавим соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
df['living_area_ratio'] = round(df['living_area'] / df['total_area'], 2)
df['kitchen_area_ratio'] = round(df['kitchen_area'] / df['total_area'], 2)
```

### Уберите редкие и выбивающиеся значения.  

На предыдущем этапе изучения датасета мы обнаружили редкие и выбивающиеся значения. 

Например, объект недвижимости площадью 900м2, что в 17 раз больше медианного значения и почти в 8 раз больше верхнего предела нормальных значений.  
Была обнаружена квартира площадью 109м2 в Санкт-Петербурге с указанной ценой 12190руб и ценой за метр 112руб. Это была ошибка и мы исправили данные. Квартиры стоимостью 12 млн оказались аномально дорогими в нашей выборке.  
Также мы изучили, что продаются квартиры с 19, 16, 15 комнатами. Их мало, но они реальны.  
В нашем исследовании предложения с количеством комнат больше 6 являются выбивающимися значениями.  
Средняя высота потолков на рынке недвижимости в Санкт-Петербурге и соседних населённых пунктах 2.55 - 2.75 м, а всё, что ниже или выше этих значений тоже является аномальным.


Чтобы убрать редкие и выбивающиеся значения нам нужно сделать еще несколько шагов: определить нормальные значения, добавить в исходный датасет соотвветствующие данные и отфильтровав по ним создать новую таблицу без аномалий.  

Для начала напишем функцию, которая будет определять границы "усов" интересующего нас параметра квартиры

```python
def whiskers(column):
    table = df[column].describe()
    
    #Нормальные значения составляют межквартирльный размах (строки 25% и 75%):
    normal_values = table[6] - table[4]
    
    #Левый и правый ус значений изучаемого параметра: 
    left_whisker = round(table[4] - 1.5 * normal_values, 2)
    right_whisker = round(table[6] + 1.5 * normal_values, 2)

    #Левый ус не должен быть меньше минимального значения:
    if left_whisker < table[3]: left_whisker = table[3] 
        
    #Правый ус не должен быть больше максимального значения:
    if right_whisker > table[7]: right_whisker = table[7]
        
    return [left_whisker, right_whisker]
```

Затем мы добавим столбцы характеризующие является параметр нормальным или аномальным.

```python
df['anomalous_area'] = df['total_area'] > whiskers('total_area')[1]
df['anomalous_price'] = df['last_price'] > whiskers('last_price')[1]
df['anomalous_rooms'] = df['rooms'] > whiskers('rooms')[1]
df['anomalous_ceiling'] = (df['ceiling_height'] > whiskers('ceiling_height')[1]) | (df['ceiling_height'] < whiskers('ceiling_height')[0])
df['too_slow'] = df['days_exposition'] > whiskers('days_exposition')[1]
df['too_fast'] = df['days_exposition'] < 45 #45 дней мы решили считать быстрой продажей
```

```python
df.head()
```

Cоздадим новый датасет, удалив из старого редкие и выбивающиеся значения. Используем параметр drop, чтобы избежать добавления старого индекса в качестве столбца

```python
normal_df = df[(df['anomalous_area'] == False) & 
     (df['anomalous_price'] == False) &
     (df['anomalous_rooms'] == False) &
     (df['anomalous_ceiling'] == False) &
     (df['too_slow'] == False) &
     (df['too_fast'] == False)].reset_index(drop=True)
```
